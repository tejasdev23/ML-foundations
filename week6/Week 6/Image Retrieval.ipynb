```python
dog_data=image_train[image_train['label']=='dog']
cat_data=image_train[image_train['label']=='cat']
car_data=image_train[image_train['label']=='automobile']
bird_data=image_train[image_train['label']=='bird']
```

```python
dog_model=graphlab.nearest_neighbors.create(dog_data,features=['deep_features'],label='id')
cat_model=graphlab.nearest_neighbors.create(cat_data,features=['deep_features'],label='id')
car_model=graphlab.nearest_neighbors.create(car_data,features=['deep_features'],label='id')
bird_model=graphlab.nearest_neighbors.create(bird_data,features=['deep_features'],label='id')
```

```python
dog_test=image_test[image_test['label']=='dog']
cat_test=image_test[image_test['label']=='cat']
car_test=image_test[image_test['label']=='automobile']
bird_test=image_test[image_test['label']=='bird']
```

```python
get_image_from_id(dog_model.query(dog_test[0:1]))['image'].show()
```

```python
cat_model.query(cat_test[0:1])['distance'].mean()
```

```python
dog_dog=dog_model.query(dog_test,k=1)['distance']
dog_cat=cat_model.query(dog_test,k=1)['distance']
dog_car=car_model.query(dog_test,k=1)['distance']
dog_bird=bird_model.query(dog_test,k=1)['distance']
```

```python
Combined_dist=graphlab.SFrame({'dog-dog':dog_dog,'dog-cat':dog_cat,'dog-car':dog_car,'dog-bird':dog_bird})
```

```python
Combined_dist
```

```python
Combined_dist['dog-dog'][0]
```

```python
def is_dog_correct(row):
    if(row['dog-dog']<row['dog-bird'] and row['dog-dog']<row['dog-cat'] and row['dog-dog']<row['dog-car']):
        return 1
    else:
        return 0
```

```python
onezerokiduniya=Combined_dist.apply(is_dog_correct)
onezerokiduniya[0]
```
